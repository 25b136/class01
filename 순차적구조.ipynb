```python
# prompt: 로봇청소기인데 적당한 속도로 움직이게해줘그리고 움직인뒤에  흔적이 남게

import matplotlib.pyplot as plt
import matplotlib.animation as animation
import random
import numpy as np

# Set up the plot
fig, ax = plt.subplots()
ax.set_xlim(0, 10)
ax.set_ylim(0, 10)
ax.set_aspect('equal', adjustable='box')
ax.set_title("Robotic Vacuum Cleaner Simulation")
ax.set_xlabel("X-coordinate")
ax.set_ylabel("Y-coordinate")
ax.grid(True)

# Initial position of the robot
x, y = 5, 5

# Store the trail
trail_x = [x]
trail_y = [y]

# Plot the initial position and the trail
robot_plot, = ax.plot(x, y, 'ro', markersize=10, label='Robot')
trail_plot, = ax.plot(trail_x, trail_y, '-', color='blue', linewidth=2, label='Trail')

# Add a legend
ax.legend()

def update(frame):
  """Updates the position of the robot and the trail."""
  global x, y, trail_x, trail_y

  # Move in a random direction with a small step size (adjust for speed)
  angle = random.uniform(0, 2 * np.pi)
  step_size = 0.2 # Adjust this value for the desired speed
  dx = step_size * np.cos(angle)
  dy = step_size * np.sin(angle)

  # Update position, keeping within bounds
  x = max(0, min(10, x + dx))
  y = max(0, min(10, y + dy))

  # Add current position to the trail
  trail_x.append(x)
  trail_y.append(y)

  # Update the plot data
  robot_plot.set_data(x, y)
  trail_plot.set_data(trail_x, trail_y)

  return robot_plot, trail_plot

# Create the animation
# interval controls the speed of the animation in milliseconds
ani = animation.FuncAnimation(fig, update, frames=200, interval=50, blit=True)

plt.show()
```
